```python
import pandas as pd 
import numpy as np 
```

```python
train_data = pd.read_csv('modified_dataset.csv',index_col=0)
train_data
```

```python
X = train_data.drop('Price',axis=1)
y = train_data['Price']
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV,KFold,RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

kfold = KFold(n_splits=5, random_state=42,shuffle=True)
```

```python
lr = LinearRegression()

result = cross_val_score(lr, X, y, cv=kfold, 
                          scoring='r2')
print(result,result.mean())
```

```python
svr = SVR(kernel='linear')

result = cross_val_score(svr, X, y, cv=kfold, 
                          scoring='r2')
print(result,result.mean())
```

```python
svr = SVR(kernel='rbf')

result = cross_val_score(svr, X, y, cv=kfold, 
                          scoring='r2')
print(result,result.mean())
```

## Random Forest

```python
ran = RandomForestRegressor()

result = cross_val_score(ran, X, y, cv=kfold, 
                          scoring='r2')
print(result,result.mean())
```

```python
parameters = {'n_estimators':np.arange(30,150,30),
              'min_samples_split':np.arange(20,150),
              'min_samples_leaf':np.arange(5,100),
              'max_features': np.arange(1,30),
              'max_depth':np.arange(10,25)}

model_rf = RandomForestRegressor(random_state=42)
cv = RandomizedSearchCV(model_rf, param_distributions=parameters,
                  cv=kfold,scoring='r2',n_iter=60)

cv.fit( X , y )

results_df = pd.DataFrame(cv.cv_results_)

print(cv.best_params_)

print(cv.best_score_)

print(cv.best_estimator_)
```

```python
model_rf = RandomForestRegressor(random_state=42,
                                 n_estimators=cv.best_params_['n_estimators'],
                                 max_features=cv.best_params_['max_features'],\
                                 min_samples_leaf=cv.best_params_['min_samples_leaf'],\
                                 max_depth= cv.best_params_['max_depth'],\
                                 min_samples_split = cv.best_params_['min_samples_split'],\
                                 oob_score=True)

results = cross_val_score(model_rf, X, y, cv=kfold, 
                          scoring='r2')

print(results,results.mean())
```

# XGBOOST

```python
xgb = XGBRegressor()
result = cross_val_score(xgb, X, y, cv=kfold, 
                          scoring='r2')
print(result,result.mean())
```

```python
lr_range = [0.001, 0.01, 0.1, 0.2,0.25, 0.3]
#n_est_range = [10,20,30,50,100]
md_range = [2,4,6,8,10]

parameters = dict(learning_rate=lr_range,
#                  n_estimators=n_est_range,
                  max_depth=md_range)

from sklearn.model_selection import GridSearchCV,cross_val_score
clf = XGBRegressor(random_state=42,tree_method='gpu_hist',gpu_id=0)
cv = GridSearchCV(clf, param_grid=parameters,
                  cv=5,scoring='r2')

cv.fit(X,y)

print(cv.best_params_)

print(cv.best_score_)

clf = XGBRegressor(max_depth=cv.best_params_['max_depth'],\
                   learning_rate=cv.best_params_['learning_rate'])


results = cross_val_score(clf,X,y,cv=kfold,scoring='r2')
print(results,results.mean())
```

```python
from sklearn.ensemble import StackingRegressor

models_considered = [('XGBoost Rgressor', xgb),
                     ('Random Forest', ran)]
```

```python
xgb1 = XGBRegressor(random_state=42,max_depth=6,learning_rate=0.25)

stack = StackingRegressor(estimators = models_considered,
                           final_estimator=xgb1)

results = cross_val_score(stack,X,y,cv=kfold,scoring='r2')
print(results,results.mean())
```

```python
xgb2 = XGBRegressor(random_state=42,max_depth=6,learning_rate=0.25)

stack = StackingRegressor(estimators = models_considered,
                           final_estimator=xgb2,passthrough=True)

results = cross_val_score(stack,X,y,cv=kfold,scoring='r2')
print(results,results.mean())
```

 Among all XGBoost Regressor without tuning has given better results
